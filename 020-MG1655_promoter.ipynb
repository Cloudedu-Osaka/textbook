```python
s = orfipy_core.orfs(s0s0rc)
sorted(s,key=lambda x: int(x[3].split(";")[2].split("=")[1]), reverse=True)
```

```python
def protain(seq_orf, s_offset=0, reverse=370000):
    st = s[seq_orf-1][0]+s_offset
    ed = s[seq_orf-1][1]
    pr = 160
    gn = s0s0rc[st-pr:ed]
    sc = s[seq_orf-1][3].split(";")[4].split(":")[1]
    
    for i in range(0,pr+10):
        print("{:> 8d} ".format(reverse-(st-pr)-i), end="")
        print("{:> 4d} ".format(i-pr), end="")
        print(gn[i:i+6],end="")
        print(" {:0.4f}".format(difflib.SequenceMatcher(None, "TTGACA", gn[i:i+6]).ratio()), end="")
        print(" {:0.4f}".format(difflib.SequenceMatcher(None, "TATAAT", gn[i:i+6]).ratio()), end="")
        #print(" {:0.4f}".format(difflib.SequenceMatcher(None, s0s0rc[st:st+3],  gn[i:i+3]).ratio()))
        print()

    return sc, Seq(sc).translate(), s0s0rc[st:st+3], Seq(s0s0rc[st:ed+3]).translate()
```

```python
# mhpR, Position relative to start of first gene (mhpR): -116
protain(126, s_offset=114)
```
